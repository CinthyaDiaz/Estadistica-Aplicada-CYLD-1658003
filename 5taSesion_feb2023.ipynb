```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#dataframe

mun_nl = pd.DataFrame({'municipio': ['Monterrey', 'Guadalupe', 'Apodaca', 'San Nicolas'], 'nacimientos':[500, 630, 768, 400]})
```

```python
#grafica simple
plt.bar(mun_nl['municipio'], ##eje x
        mun_nl['nacimientos'], ##eje y
       color=['blue','green','red','yellow'])
#agregar titulo
plt.title('Nacimientos por municipio')

#agregar nombre al eje x
plt.xlabel('Municipios de NL')

#agregar nombre al eje y
plt.ylabel('Cantidad de nacimientos')

#cambiar los saltos escala y
plt.yticks([0,200,400,600,800])
plt.show()
```

```python
#ordenamos el dataframe
mun_nl = mun_nl.sort_values('nacimientos', ascending=True)

#graficamos
plt.bar(mun_nl['municipio'], ##eje x
        mun_nl['nacimientos'], ##eje y
       color=['blue','green','red','yellow']) ##color de las barras
#agregar titulo
plt.title('Nacimientos por municipio')

#agregar nombre al eje x
plt.xlabel('Municipios de NL')

#agregar nombre al eje y
plt.ylabel('Cantidad de nacimientos')

#cambiar los saltos escala y
plt.yticks([0,200,400,600,800])
plt.show()
```

```python
#fig es nombre del lienzo
fig, (panel1, panel2) = plt.subplots(2,1) #subgraficas, nombre y acom
fig.tight_layout()

#Panel1 (subgrafica 1)
x1 = x #ejex de grafico 1
y1 = y #eje y de grafico 1
panel1.plot(x,y,label='nombre de línea')
panel1.title.set_text('Grafica 1')

#Panel2 (subgrafica 1)
x2 = mun_nl['municipio'] #eje x de grafico 2
y2 = mun_nl['nacimientos'] #eje y de grafico 2
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show()
```

```python
#fig es nombre del lienzo
fig, (panel1, panel2) = plt.subplots(1,2) #subgraficas, nombre y acom
fig.tight_layout()

#Panel1 (subgrafica 1)
x1 = x #ejex de grafico 1
y1 = y #eje y de grafico 1
panel1.plot(x,y,label='nombre de línea')
panel1.title.set_text('Grafica 1')

#Panel2 (subgrafica 1)
x2 = mun_nl['municipio'] #eje x de grafico 2
y2 = mun_nl['nacimientos'] #eje y de grafico 2
panel2.bar(x2,y2)
panel2.title.set_text('Grafica 2')

plt.show()
```

```python
mun_nl = mun_nl.sort_values('nacimientos', ascending=True)
plt.barh(mun_nl['municipio'], mun_nl['nacimientos'], color=['blue','green','red','yellow'])
plt.title('Nacimientos por municipio')
plt.xlabel('Municipios de NL')
plt.ylabel('Cantidad de nacimientos')
plt.show()
```
