1. El banco sospecha que entre mas pasa el tiempo más díficil es que puedan pagar los clientes que se les concedió un préstamo. 

a) Calcula el promedio del monto del préstamo de los que pagaron y el promedio del monto de los que no han pagado.

b) Ajusta un modelo weibull y un modelo log-normal y graficas las funciones de riesgo respectivas. 


```python
Pagaron = datos[datos['full_repaid'] == 1]
Prompag = Pagaron['amount'].mean()

NoPagaron = datos[datos['full_repaid'] == 0]
Promnp = NoPagaron['amount'].mean()

print('Monto promedio del prestamo de los que pagaron: ',Prompag, 'Monto promedio del prestamo de los que no pagaron: ', Promnp)
```

**Modelo Weibull**

```python
from lifelines import WeibullFitter
```

```python
Weib = WeibullFitter(label='Duración del Prestamo')
Weib.fit(datos['duration'],datos['full_repaid'])
```

```python
Weibpag = WeibullFitter(label='Duración del Prestamo de los que pagaron')
Weibpag.fit(Pagaron['duration'],Pagaron['full_repaid'])
Weibnp = WeibullFitter(label='Duración del Prestamo de los que no pagaron')
Weibnp.fit(NoPagaron['duration'],NoPagaron['full_repaid'])
```

```python
Weib.plot_hazard();
```

**Modelo Log-Normal**

```python
from lifelines import LogNormalFitter
```

```python
Log = LogNormalFitter(label='Duración del Prestamo')
Log.fit(datos['duration'],datos['full_repaid'])
```

```python
Logpag = LogNormalFitter(label='Duración del Prestamo de los que pagaron')
Logpag.fit(Pagaron['duration'],Pagaron['full_repaid'])
Lognp = LogNormalFitter(label='Duración del Prestamo de los que no pagaron')
Lognp.fit(NoPagaron['duration'],NoPagaron['full_repaid'])
```

```python
Log.plot_hazard();
```

```python
import statistics
```

```python
# a)
med = datos['amount'].median()
med
```

```python
datos['Deben'] = np.where(datos['amount']>=med, 1, 0)
datos
```

```python
# b)
Dmucho = datos[datos['Deben'] == 1] 
Dpoco = datos[datos['Deben'] == 0]
```

```python
kmdm = KaplanMeierFitter(label='Deben mucho')
kmdm.fit(Dmucho['duration'],Dmucho['full_repaid'])

kmdp = KaplanMeierFitter(label='Deben poco')
kmdp.fit(Dpoco['duration'],Dpoco['full_repaid'])
```

```python
kmdm.survival_function_
```

```python
kmdp.survival_function_
```

```python
kmdm.plot()
kmdp.plot();
```
