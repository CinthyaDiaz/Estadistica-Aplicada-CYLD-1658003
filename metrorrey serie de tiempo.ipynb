```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from statsmodels.tsa.api import seasonal_decompose
```

```python
metrorrey = pd.read_csv('metrorrey.csv')
```

```python
metrorrey.head()
```

```python
metrorrey.tail()
```

```python
# vamos a definir como tiempo a la variable fecha
metrorrey['Fecha'] = pd.to_datetime(metrorrey['Fecha'], format = '%Y/%m')
```

```python
metrorrey.tail()
```

```python
# la variable de fecha tenemos que convertirla en el index
metrorrey = metrorrey.set_index('Fecha')
```

```python
# definir la frecuencia (en este caso mensual)
metrorrey.index.freq = 'MS'
```

```python
# podemos hacer cualquier análisis de series
# confirmar
metrorrey.index
```

```python
metrorrey['Pasajeros'].plot();
```

```python
metrorrey['Ingresos'].plot();
```

```python
#serie descompuesta
seried = seasonal_decompose(metrorrey['Pasajeros'], model = 'ad')
seried.plot();
```
