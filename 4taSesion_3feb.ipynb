```python
#renombrar una columna
calificaciones = calificaciones.rename(columns={'Calficación Final':'cal_final'})
```

```python
calificaciones.head() 
```

```python
#Calificacion final
calificaciones['cal_final'].mean()
```

```python
#condicional
calificaciones['situación'] = np.where(calificaciones['cal_final']>=70, 'Aprobado', 'Reprobado')
```

```python
calificaciones.head()
```

```python
#contar en grupos
calificaciones.groupby('situación').count()
```

```python
calificaciones.value_counts('situación')
```

```python
#ordenar calificaciones
calificaciones = calificaciones.sort_values('cal_final', ascending=False)
```
